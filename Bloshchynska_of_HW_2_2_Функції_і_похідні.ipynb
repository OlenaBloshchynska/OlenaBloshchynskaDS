1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt

# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = 10 * np.ones_like(x)  # Похідна f'(x) = 10

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

# Позначення точки та дотичної лінії
x_point = 3
f_point = 10 * x_point - 5
df_point = 10
tangent_line = df_point * (x - x_point) + f_point  # Рівняння дотичної: y - y1 = m(x - x1)

plt.scatter(x_point, f_point, color='red') # точка на кривій
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 3') # дотична

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Дотична збігається з лінійною функцією, оскільки для лінійної функції будь-яка дотична в будь-якій точці збігається з самою функцією. Тим не менш функція вказує, що збільшення годин впливає на підвищення оцінки.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Визначення символьних змінних
x = sp.Symbol('x')

# Визначення функцій
f1_x = sp.sin(x)
f2_x = sp.ln(x)

# Обчислення похідних
f1_prime_x = sp.diff(f1_x, x)
f2_prime_x = sp.diff(f2_x, x)

# Перетворення символьних функцій у функції Python
f1 = sp.lambdify(x, f1_x, 'numpy')
f1_prime = sp.lambdify(x, f1_prime_x, 'numpy')
f2 = sp.lambdify(x, f2_x, 'numpy')
f2_prime = sp.lambdify(x, f2_prime_x, 'numpy')

# Генерація значень x
x_values = np.linspace(0.1, 2 * np.pi, 100)

# Обчислення відповідних значень f(x) та f'(x)
y_values_f1 = f1(x_values)
y_prime_values_f1 = f1_prime(x_values)
y_values_f2 = f2(x_values)
y_prime_values_f2 = f2_prime(x_values)

# Побудова графіків
plt.figure(figsize=(12, 8))

# Графік для f(x) = sin(x)
plt.subplot(2, 2, 1)
plt.plot(x_values, y_values_f1, label='f(x) = sin(x)')
plt.plot(x_values, y_prime_values_f1, label="f'(x) = cos(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x) та її похідна')
plt.legend()

# Графік для f(x) = ln(x)
plt.subplot(2, 2, 2)
plt.plot(x_values, y_values_f2, label='f(x) = ln(x)')
plt.plot(x_values, y_prime_values_f2, label="f'(x) = 1/x")
plt.xlabel('x')
plt.ylabel('y')
plt.title('ln(x) та її похідна')
plt.legend()

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Визначення символічної змінної
x = sp.symbols('x')

# Функція загальних витрат
f = x**2 - 5*x + 1

# Обчислення похідної
df = sp.diff(f, x)

# Перетворення символічних виразів у числові функції
f_func = sp.lambdify(x, f, 'numpy')
df_func = sp.lambdify(x, df, 'numpy')

# Генерація значень x від 0 до 100
x_values = np.linspace(0, 100, 400)

# Обчислення значень функції та похідної на вказаному діапазоні
f_values = f_func(x_values)
df_values = df_func(x_values)

# Побудова графіків
plt.figure(figsize=(10, 5))
plt.plot(x_values, f_values, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x_values, df_values, label="f'(x) = 2x - 5", color='green')

# Додавання легенди, підписів та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
